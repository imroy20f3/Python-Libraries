# Bar Plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,35,45,65]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("BAR",fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",label="Popularity")
plt.legend()
plt.show()
#plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5,label="Popularity")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,35,45,65]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("BAR",fontsize=15)

plt.bar(p,y,width,color="r",label="Popularity")
plt.bar(p1,z,width,color="g",label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
# pie -> single and multiple
df = pd.DataFrame(
    {
        'batsman':['Dhawan','Rohit','Kohli','SKY','Pandya','Pant'],
        'match1':[120,90,35,45,12,10],
        'match2':[0,1,123,130,34,45],
        'match3':[50,24,145,45,10,90]
    }
)

df.head()
```

```python
df['match1'].plot(kind='pie',labels=df['batsman'].values,autopct='%0.1f%%')
```
