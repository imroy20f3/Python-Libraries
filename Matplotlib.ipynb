# Bar Plot

```python
import matplotlib.pyplot as plt
x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,35,45,65]
plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("BAR",fontsize=15)
c=["y","b","m","g"]
plt.bar(x,y,width=0.4,color=c,edgecolor="r",label="Popularity")
plt.legend()
plt.show()
#plt.bar(x,y,width=0.4,color=c,edgecolor="r",linewidth=5,linestyle=":",alpha=0.5,label="Popularity")
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,35,45,65]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("BAR",fontsize=15)

plt.bar(p,y,width,color="r",label="Popularity")
plt.bar(p1,z,width,color="g",label="Popularity1")

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt


x=["python","c","c++","java"]
y=[85,70,60,82]
z=[20,35,45,65]

plt.xlabel("language",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.title("BAR",fontsize=15)

plt.barh(x,y,color="r",label="Popularity")
plt.barh(x,z,color="g",label="Popularity1")

plt.legend()
plt.show()
```

# Scatter plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,6,7]
plt.scatter(day,no,color="r")
plt.title("scatter",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,6,7]
colors=["r","y","g","b","r","g","r"]
sizes=[100,50,40,30,100,90,70]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.4,marker="*",edgecolor="g")
plt.title("scatter",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,6,7]
c=["r","y","g","b","r","g","r"]
sizes=[100,50,40,30,100,90,70]
plt.scatter(day,no,color=c,s=sizes,marker="*",edgecolor="g",linewidth=2)
plt.title("scatter",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,6,7]
colors=[10,29,30,49,56,20,30]
sizes=[100,50,40,30,100,90,70]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis",alpha=0.5)
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("scatter",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,6,7]
no2=[3,2,4,5,6,9,8]
colors=[10,29,30,49,56,20,30]
sizes=[100,50,40,30,100,90,70]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis",alpha=0.5)
plt.scatter(day,no2,color="r",s=sizes)
t=plt.colorbar()
t.set_label("colorbar",fontsize=15)
plt.title("scatter",fontsize=15)
plt.xlabel("Day",fontsize=15)
plt.ylabel("No.",fontsize=15)
plt.show()
```

# pie chart

```python
import matplotlib.pyplot as plt

x=[10,20,30,40]
y=["c","c++","java","pyton"]
ex=[0.4,0.0,0.0,0.0]
c=["r","g","b","y"]

plt.pie(x,labels=y,explode=ex,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.1,startangle=0
       ,textprops={"fontsize":15},counterclock=False,wedgeprops={"linewidth":5},rotatelabels=False)

plt.title("Pie Chart")
plt.legend(loc=2)
plt.show()
```

# Stem Plot

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,5,2,6,4,3]

plt.stem(x,y,linefmt=":",markerfmt="ro",bottom=0,basefmt="g",label="Python",use_line_collection=False)

plt.legend()
plt.show()
```
